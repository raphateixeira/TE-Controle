```python
import control as ct
import matplotlib.pyplot as plt
import numpy as np
```

```python
m = 1.2
b = 0.4
k = 0.1

G = ct.TransferFunction(1,[m,b,k])
```

```python
ct.poles(G)
```

```python
ct.pzmap(G)
```

```python
t,y = ct.step_response(G)
```

```python
plt.figure(figsize = (6,3))
plt.plot(t,y)
plt.xlabel('t')
plt.ylabel('y')
plt.grid()
plt.show()
```

```python
wn = np.sqrt(k/m)
print(wn)
```

```python
qsi = b/(2*m*wn)
print(qsi)
```

```python
ts = 4/(qsi*wn)
print(ts)
```

```python
OS = np.exp(-np.pi/np.sqrt(1-qsi**2))*100
print(OS)
```
