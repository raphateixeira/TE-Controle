```python
import numpy as np
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
from sklearn.preprocessing import MinMaxScaler
import matplotlib.pyplot as plt
```

```python
A = 5.0
a = 0.2
w = 2.0

def funcao(x):
    return A * np.exp(-a * x) * np.cos(w * x)
```

```python
num_samples_train = 1000
x_train_raw = np.random.uniform(0, 15, (num_samples_train, 1))
y_train_raw = funcao(x_train_raw)

num_samples_test = 200
x_test_raw = np.random.uniform(0, 15, (num_samples_test, 1))
y_test_raw = funcao(x_test_raw)
```

```python
input_scaler = MinMaxScaler()
output_scaler = MinMaxScaler()

x_train_scaled = input_scaler.fit_transform(x_train_raw)
y_train_scaled = output_scaler.fit_transform(y_train_raw)

x_test_scaled = input_scaler.transform(x_test_raw)
y_test_scaled = output_scaler.transform(y_test_raw)
```

```python
model = keras.Sequential(
    [
        layers.Input(shape=(1,), name="input_layer"),
        layers.Dense(64, activation="relu", name="hidden_layer_1"),
        layers.Dense(64, activation="relu", name="hidden_layer_2"),
        layers.Dense(32, activation="relu", name="hidden_layer_3"),
        layers.Dense(1, name="output_layer"),
    ]
)
```

```python
model.summary()
```

```python
model.compile(
    optimizer=keras.optimizers.Adam(learning_rate=0.001),
    loss="mean_squared_error",
)

print("\nIniciando o treinamento...")

history = model.fit(
    x_train_scaled,
    y_train_scaled,
    epochs=500,
    batch_size=32,
    validation_data=(x_test_scaled, y_test_scaled),
    verbose=1
)

print("Treinamento concluído.")

plt.figure(figsize=(14, 6))
plt.subplot(1, 2, 1)
plt.plot(history.history['loss'], label='Perda de Treinamento')
plt.plot(history.history['val_loss'], label='Perda de Validação')
plt.title('Histórico de Perda (Loss) do Modelo')
plt.xlabel('Época')
plt.ylabel('Erro Quadrático Médio (MSE)')
plt.legend()

x_plot = np.linspace(0, 15, 500).reshape(-1, 1)
y_plot_true = funcao(x_plot)

x_plot_scaled = input_scaler.transform(x_plot)
y_plot_pred_scaled = model.predict(x_plot_scaled)
y_plot_pred = output_scaler.inverse_transform(y_plot_pred_scaled)

plt.subplot(1, 2, 2)
plt.scatter(x_train_raw, y_train_raw, alpha=0.1, label='Dados de Treino (Amostras)')
plt.plot(x_plot, y_plot_true, 'r-', linewidth=3, label='Função Real $f(x)$')
plt.plot(x_plot, y_plot_pred, 'g--', linewidth=2, label='Previsão da Rede Neural')
plt.title('Aproximação da Função')
plt.xlabel('$x$')
plt.ylabel('$f(x)$')
plt.legend()
plt.grid(True)
plt.tight_layout()
```
