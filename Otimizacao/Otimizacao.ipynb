```python
import numpy as np
import matplotlib.pyplot as plt
from  scipy.optimize import minimize
```

```python
# 1. Definir a função
f1 = lambda x: x**2 +3*x- 1

# 2. Minimizar a função
x0_1 = 1.0 # Chute inicial
sol1 = minimize(f1, x0_1, method='nelder-mead') # 'nelder-mead' é um método robusto para funções simples

# 3. Plotar o gráfico e a solução
x_range = np.linspace(-2, 2, 500)
plt.figure(figsize=(8, 6))
plt.plot(x_range, f1(x_range), label='$f(x) = x^2 + 1$', color='blue')

# Adiciona o ponto de solução
# sol1.x é o valor de x que minimiza a função
# sol1.fun é o valor da função no ponto de mínimo
plt.scatter([sol1.x], [sol1.fun], color='red', marker='o', s=100, label=f'Mínimo: x={sol1.x[0]:.4f}, f(x)={sol1.fun:.4f}')

plt.title('Minimização da Função $f(x) = x^2 + 1$')
plt.xlabel('$x$')
plt.ylabel('$f(x)$')
plt.grid(True)
plt.legend()
plt.show()

print(f"Solução para f(x) = x**2 + 1:")
print(f"  Mínimo encontrado em x = {sol1.x[0]:.6f}")
print(f"  Valor da função no mínimo = {sol1.fun:.6f}")
```

```python
# 1. Definir a segunda função
f2 = lambda x : (x - 2)**2 * (x + 2)**2 - x

# 2. Minimizar a função
x0_2 = 0.0 # Chute inicial
# 'BFGS' é um algoritmo mais adequado para funções não quadráticas e de múltiplas variáveis, mas funciona bem aqui
sol2 = minimize(f2, x0_2, method='BFGS')

# 3. Plotar o gráfico e a solução
x_range_2 = np.linspace(-3, 3, 500) # Ajusta o intervalo para visualizar melhor os mínimos
plt.figure(figsize=(8, 6))
plt.plot(x_range_2, f2(x_range_2), label='$f(x) = (x - 2)^2 (x + 2)^2 - x$', color='darkgreen')

# Adiciona o ponto de solução
plt.scatter([sol2.x], [sol2.fun], color='red', marker='o', s=100, label=f'Mínimo (próximo de $x_0=0$): x={sol2.x[0]:.4f}, f(x)={sol2.fun:.4f}')

plt.title('Minimização da Função $f(x) = (x - 2)^2 (x + 2)^2 - x$')
plt.xlabel('$x$')
plt.ylabel('$f(x)$')
plt.grid(True)
plt.legend()
plt.show()

print(f"Solução para f(x) = (x - 2)**2 * (x + 2)**2 - x (com x0 = {x0_2}):")
print(f"  Mínimo local encontrado em x = {sol2.x[0]:.6f}")
print(f"  Valor da função no mínimo = {sol2.fun:.6f}")
```
